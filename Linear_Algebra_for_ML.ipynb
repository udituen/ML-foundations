 Variance Inflation Factor (VIF) [Most Common]
What it does: Measures how much the variance of a regression coefficient is inflated due to multicollinearity.

Rule of Thumb:

VIF  1  No multicollinearity.

VIF > 5 (sometimes > 10)  Serious multicollinearity (linear dependence is likely).

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# X should be your features as a DataFrame, not including the target column
vif_data = pd.DataFrame()
vif_data["feature"] = iris.feature_names
vif_data["VIF"] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]
print(vif_data)
```
